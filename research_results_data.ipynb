```python
import pandas as pd

path_posts_words = '/Users/lenar/PycharmProjects/vk-word-frequency-analyzer/results/frequency_post_words_in_3000_posts.csv'
path_main_comm_words = '/Users/lenar/PycharmProjects/vk-word-frequency-analyzer/results/frequency_main_comm_words_in_3000_posts.csv'
path_answ_comm_words = '/Users/lenar/PycharmProjects/vk-word-frequency-analyzer/results/frequency_answ_comm_words_in_3000_posts.csv'
```

### Загрузка данных

```python
posts_words_freq = pd.read_csv(path_posts_words, header=None, names=['word', 'frequency'])
main_comm_words_freq = pd.read_csv(path_main_comm_words, header=None, names=['word', 'frequency'])
answ_comm_words_freq = pd.read_csv(path_answ_comm_words, header=None, names=['word', 'frequency'])

posts_words_freq.head(15).T
```

#### Общая информация

```python
print(f"All len unique post words {len(posts_words_freq)}")
print(f"All len unique main comments words {len(main_comm_words_freq)}")
print(f"All len unique answer comments words {len(answ_comm_words_freq)}")
```

### Сортировка слов по частоте

```python
posts_words_freq = posts_words_freq.sort_values(by=['frequency'], ascending=False)
main_comm_words_freq = main_comm_words_freq.sort_values(by=['frequency'], ascending=False)
answ_comm_words_freq = answ_comm_words_freq.sort_values(by=['frequency'], ascending=False)
```

```python
posts_words_freq.head(15).T
```

```python
main_comm_words_freq.head(15).T
```

### Рассмотрим на наиболее частотные слова

```python
most_freq_posts_words = posts_words_freq.iloc[:40]
most_freq_main_comm_words = main_comm_words_freq.iloc[:40]
most_freq_answ_comm_words = answ_comm_words_freq.iloc[:40]
```

```python
most_freq_posts_words.plot(x = 'word', y = 'frequency', kind='bar', figsize = (10,5), 
                           grid=True, fontsize=13, legend=True, title = 'Наиболее частотные слова в постах');

most_freq_main_comm_words.plot(x = 'word', y = 'frequency', kind='bar', figsize = (10,5), 
                           grid=True, fontsize=13, legend=True, title = 'Наиболее частотные слова в главных комментариях');

most_freq_answ_comm_words.plot(x = 'word', y = 'frequency', kind='bar', figsize = (10,5), 
                           grid=True, fontsize=13, legend=True, title = 'Наиболее частотные слова в ответах на другие комментарии');
```

```python
set_most_freq_words_in_posts = set(most_freq_posts_words['word'])
set_most_freq_words_in_comment = set.union(set(most_freq_main_comm_words['word']), set(most_freq_answ_comm_words['word']))
```

#### Посмотрим слова, которые являеются часто используемыми в постах, но не в комментариях

```python
print(*(set_most_freq_words_in_posts-set_most_freq_words_in_comment))
```

#### и наоборот

```python
print(*(set_most_freq_words_in_comment-set_most_freq_words_in_posts))
```
